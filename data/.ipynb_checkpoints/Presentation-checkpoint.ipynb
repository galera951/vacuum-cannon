```python
import numpy as np
```

```python
def lessSquaresMethod(x_dots: np.array, y_dots: np.array, eps=5):
    """
    :param eps: rounding parameter
    :param x_dots: X dots
    :param y_dots: Y dots
    :return: Line approximation coefficients with less squares method
    """
    x_dots = np.array(x_dots)
    y_dots = np.array(y_dots)

    if len(x_dots) != len(y_dots):
        raise f"Count of x_dots ({len(x_dots)}) not equal to count of y_dots ({len(y_dots)})"

    n = len(x_dots)

    xy = np.sum(x_dots * y_dots) / n
    x_av = np.sum(x_dots) / n
    y_av = np.sum(y_dots) / n
    x2_av = np.sum(x_dots ** 2) / n

    k = round((xy - x_av * y_av) / (x2_av - x_av ** 2), eps)
    b = round((y_av - k * x_av), eps)

    # print(f"y = {k} * x + {b}")
    
    sigma = np.sqrt(sum([x_dots[i] - x_av for i in range(n)]) / (n - 1)) # Случайная ошибка
    
    return k, b, sigma
```

## Analysis of experimental data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sympy as smp
```

```python
data = pd.read_csv('data.csv')
```

$v^2(H)$

```python
def torrich2(h):
    return 2 * 9.802 * h
```

```python
H = data['H'] / 1000 # метры
t = data['t']        # секунды
delta_h = data['delta_h'] / 1000 # метры
Q = 0.19 * 0.18 * delta_h / t # метры ** 3 / секунду
delta_Q = np.sqrt((0.001 / 0.19) ** 2 * 2 + (0.001 / delta_h) ** 2 + (0.5 / t) ** 2)
v2_exp = 4 * Q / (np.pi * (0.01 ** 2))
delta_v2_exp = np.sqrt(delta_Q ** 2 + 0.001 / 0.01)
v2_torr = torrich2(H) / 10
```

```python
k, b, sigma = lessSquaresMethod(H, v2_exp)
sigma
```

```python
plt.figure(figsize=(5, 3.5), dpi=250)
plt.plot(H, v2_torr, '-g', label='Torricelli')
plt.plot(H, v2_torr, '.g')
plt.plot(H, k * H + b, '-r', label='Experiment')
plt.errorbar(H, v2_exp, xerr=0.003, yerr=v2_exp * delta_v2_exp, fmt='.r')

plt.xlabel('$H, m$')
plt.ylabel('$v^2, m^2/s^2$')
plt.title('Dependence of velocity square on height', size='medium')
plt.legend()
plt.tight_layout()
plt.savefig(r'LyX-layout/img/v2H.png')
```

```python
def ventury(p1, p2):
    rho = 1000
    S1 = np.pi * (0.01 / 2) ** 2
    S2 = np.pi * (0.006 / 2) ** 2
    return np.sqrt((2 * (p1 - p2)) / (rho * ((S1 / S2) ** 2 - 1)))
```

```python
def pitot(p1, p2):
    rho = 1000
    return (2 * (p2 - p1) / rho) ** 0.5
```

```python
def h2p(h):
    rho = 1000
    g = 9.802
    return rho * g * h
```

```python
p1 = h2p(data['vent_h1'] / 1000)
p2 = h2p(data['vent_h2'] / 1000)
sigma_p_V = np.sqrt((0.001 / data['vent_h1']) ** 2 + (0.001 / data['vent_h2']) ** 2)
v_V = ventury(p1, p2)
```

```python
p1 = h2p(data['pito_h1'] / 1000)
p2 = h2p(data['pito_h2'] / 1000)
sigma_p_P = np.sqrt((0.001 / data['pito_h1']) ** 2 + (0.001 / data['pito_h2']) ** 2)
v_P = pitot(p1, p2)
```

```python
v_exp = v2_exp ** 0.5
```

```python
plt.figure(figsize=(5, 3.5), dpi=250)

k, b, sigma = lessSquaresMethod(v_exp, v_V)
plt.errorbar(v_exp, v_V, fmt='.r', xerr=v_exp * 0.05, yerr=v_V * sigma_p_V ** 0.2)
plt.plot(v_exp, k * v_exp + b, '-r', label='Ventury')

k, b, sigma = lessSquaresMethod(v_exp, v_P)
plt.errorbar(v_exp, v_P, fmt='.g', xerr=0.05*v_exp, yerr=v_P * sigma_p_P ** 0.18)
plt.plot(v_exp, k * v_exp + b, '-g', label='Pitot')

plt.plot(v_exp, v_exp, linestyle='--', color='0.5', label='Equal')

plt.xlabel('$v_Q, m/s$')
plt.ylabel('$v, m/s$')
plt.title('Dependence of velocity for rate on velocity of Ventury and Pitot', size='medium')
plt.legend()
plt.tight_layout()
plt.savefig(r'LyX-layout/img/vVP_vR.png')
```
